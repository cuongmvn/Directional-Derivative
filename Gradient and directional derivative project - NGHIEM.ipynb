```python
import numpy as np


def inp_num(a):
    try:
        if a == 3:
            s = np.random.uniform(-5,5)
            print('You put input wrong too many times, random input',s,'is chosen')
            return s
        else:
            s = input()
            return float(s)
    except ValueError:
        print('Error: input must be numerical value! You have ', 2-a, 'try left.')
        return inp_num(a+1)


def calculateGradient(x,y):
    return 3*x*x*y-y**3, x**3-3*y*y*x


def normalize(x,y):
    z = (x**2+y**2)**0.5
    return x/z,y/z


def main():
    print("Please enter a number value for x:")
    x = inp_num(0)
    print("Please enter a number value for y:")
    y = inp_num(0)
    
    
    g1,g2 = calculateGradient(x,y)
    print("\nGradient at the point of x and y is: ", g1,g2)
    
    print("\nNow let's put in a vector to calculate directional derivative:")
    print("Please enter u[1]:")
    u1 = inp_num(0)
    print("Please enter u[2]:")
    u2 = inp_num(0)
    
    
    vector = normalize(u1,u2)
    print("\nNormalized vector is: ", vector)
    grad = (g1,g2)
    DD = np.dot(grad, vector)
    print("\nDirectional Derivative: ", DD)
    
    
if __name__ == '__main__':
    main()
```

# Bonus: Some demonstration:

```python
%matplotlib qt
import matplotlib.pyplot as plt
from matplotlib.patches import FancyArrowPatch, ArrowStyle
feature_x = np.arange(-30, 30, 1)
feature_y = np.arange(-30, 30, 1)

x, y = np.meshgrid(feature_x, feature_y)
z = x**3*y -y**3*x
u = 3*x*x*y-y**3
v = x**3-3*y*y*x

# Normalize all gradients to focus on the direction not the magnitude
norm = np.linalg.norm(np.array((u, v)), axis=0)
u = u / norm
v = v / norm

fig, ax = plt.subplots(1, 1)
ax.set_aspect(1)
ax.plot(feature_x, feature_y, c='k')
ax.quiver(x, y, u, v, units='xy', scale=1, color='gray')
ax.contour(x, y, z, 20, cmap='jet')

####We choose a point p to draw it's gradient step
x,y = 5,10
p = np.array([x,y])
for i in range(10):
    q = np.array(calculateGradient(p[0],p[1]))
    q = q/np.linalg.norm(q)
    p2 = p + q + q

    arrow = FancyArrowPatch((p[0], p[1]), (p2[0], p2[1]), arrowstyle='simple',
                        color='r', mutation_scale=5)  
    ax.add_patch(arrow)
    p = p2
    

p = np.array([x,y])
for i in range(10):
    q = np.array(calculateGradient(p[0],p[1]))
    q = q/np.linalg.norm(q)
    p2 = p - q - q

    arrow = FancyArrowPatch((p[0], p[1]), (p2[0], p2[1]), arrowstyle='simple',
                        color='b', mutation_scale=5)  
    ax.add_patch(arrow)
    p = p2
```
